# AutoGrad
AutoGrad is core components of pytorch that provides automatic differentiation for tensor.

```python
import torch
```

```python
x = torch.tensor(3., requires_grad=True)
```

```python
y = x**2
y
```

```python
y.backward()
```

```python
x.grad
```

```python
# Example 2
# y=x^2, z=sin(y)
```

```python
x = torch.tensor(4., requires_grad=True)
y = x**2
z = torch.sin(y)
x,y,z
```

```python
z.backward()
```

```python
x.grad
```

```python
# Real life example
x = torch.tensor(6.7)
y = torch.tensor(0.)
```

```python
w = torch.tensor(1., requires_grad=True)
b = torch.tensor(0., requires_grad=True)
```

```python
z = w * x + b
```

```python
y_pred = torch.sigmoid(z)
```

```python
y_pred.backward()
```

```python
w.grad, b.grad
```

```python
x = torch.tensor([1.,2.,3.], requires_grad=True)
```

```python
y = (x**2).mean()
y
```

```python
y.backward()
```

```python
x.grad
```

```python
# clearing grad
x = torch.tensor(2., requires_grad=True)
```

```python
y = x**2
y
```

```python
y.backward()
```

```python
x.grad
```

```python
x.grad.zero_()
```

```python
# disable gradient tracking
# option - 1:  requires_grad(False)
x.requires_grad_(False)
```

```python
# option - 2: detach()
x = torch.tensor(2., requires_grad=True)
```

```python
z = x.detach()
z
```

```python
# option - 3: torch.no_grad()
with torch.no_grad():
    y = x**2
y
```
